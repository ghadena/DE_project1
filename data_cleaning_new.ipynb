```python
import pandas as pd 
df_spending = pd.read_csv('DE1_P1_Healthcare(healthcare_spending).csv')
df_spending_melted = pd.melt(df_spending, 
                    id_vars=['Country Name', 'Country Code', 'Series Name', 'Series Code'],  # Columns to keep
                    var_name='Year',  
                    value_name='Value')

df_spending_melted['Year'] = df_spending_melted['Year'].str.extract(r'(\d{4})')
df_spending_melted.head(20)

df_spending_melted['Value'] = pd.to_numeric(df_spending_melted['Value'], errors='coerce')
df_spending_pivoted = df_spending_melted.pivot_table(index=['Country Name', 'Country Code', 'Year'],  # These remain as rows
                                    columns='Series Name',  # Series Name becomes columns
                                    values='Value',  # Values to fill in the new table
                                    aggfunc='first')  # Use 'first' to handle duplicate entries

df_spending_pivoted = df_spending_pivoted.reset_index()


df_spending = df_spending_pivoted

df_spending['Country Name'] = df_spending['Country Name'].replace({
    'Congo, Dem. Rep.': 'Democratic Republic of Congo',
    'Gambia, The': 'Gambia', 
    "Korea, Dem. People's Rep.": "North Korea", 
    'Yemen, Rep.': 'Yemen'}, regex=True)

df_regions = pd.read_csv('DE1_P1_Healthcare(Country - Metadata).csv', encoding='ISO-8859-1')
df_spending.rename(columns={'Country Code': 'CountryCode'}, inplace=True)
df_regions.rename(columns={'Code': 'CountryCode'}, inplace=True)
df_spending_with_region = df_spending.merge(df_regions[['CountryCode', 'Region']], on='CountryCode', how='left')

```
